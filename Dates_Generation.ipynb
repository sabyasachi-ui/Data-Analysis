```python
from datetime import datetime, timedelta
import pandas as pd

# Define date range
start_date = datetime(2025, 7, 1)
end_date = datetime(2025, 8, 31)

# Allowed weekdays: Monday (0), Tuesday (1), Friday (4)
allowed_weekdays = {0, 1, 4}

# Dates to exclude
excluded_dates_str = [
    "16-08-2025", "27-08-2025", "29-09-2025", "30-09-2025", "01-10-2025",
    "02-10-2025", "21-10-2025", "25-12-2025", "01-01-2026", "14-01-2026",
    "03-03-2026", "21-03-2026", "03-04-2026", "14-04-2026", "15-06-2026",
    "15-08-2025", "26-01-2026", "01-04-2026"
]
excluded_dates = {datetime.strptime(d, "%d-%m-%Y").date() for d in excluded_dates_str}

# Generate valid dates
valid_data = []
current_date = start_date
while current_date <= end_date:
    if current_date.weekday() in allowed_weekdays and current_date.date() not in excluded_dates:
        valid_data.append({
            "Date": current_date.strftime("%d-%m-%Y"),
            "Weekday": current_date.strftime("%A")  # Full weekday name
        })
    current_date += timedelta(days=1)

# Create DataFrame and export to Excel
df = pd.DataFrame(valid_data)
df.to_excel("valid_weekdays_jul_aug_2025.xlsx", index=False)

print("✅ Excel file 'valid_weekdays_jul_aug_2025.xlsx' has been created with weekday names.")
```

```python
from datetime import datetime, timedelta
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# Define date range
start_date = datetime(2025, 7, 1)
end_date = datetime(2025, 8, 31)
allowed_weekdays = {0, 1, 4}  # Monday, Tuesday, Friday

# Excluded dates
excluded_dates_str = [
    "16-08-2025", "27-08-2025", "29-09-2025", "30-09-2025", "01-10-2025",
    "02-10-2025", "21-10-2025", "25-12-2025", "01-01-2026", "14-01-2026",
    "03-03-2026", "21-03-2026", "03-04-2026", "14-04-2026", "15-06-2026",
    "15-08-2025", "26-01-2026", "01-04-2026"
]
excluded_dates = {datetime.strptime(d, "%d-%m-%Y").date() for d in excluded_dates_str}

# Collect valid dates
valid_data = []
current = start_date
while current <= end_date:
    if current.weekday() in allowed_weekdays and current.date() not in excluded_dates:
        valid_data.append({
            "Date": current.strftime("%d-%m-%Y"),
            "Weekday": current.strftime("%A")
        })
    current += timedelta(days=1)

# Create DataFrame
df = pd.DataFrame(valid_data)

# Save to Excel
file_name = "valid_weekdays_jul_aug_2025.xlsx"
df.to_excel(file_name, index=False)

# Load Excel and apply formatting
wb = load_workbook(file_name)
ws = wb.active

# Define colors
fill_colors = {
    "Monday": "ADD8E6",   # Light Blue
    "Tuesday": "90EE90",  # Light Green
    "Friday": "FFFF99",   # Light Yellow
}

# Apply color formatting
for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=2, max_col=2):
    for cell in row:
        weekday = cell.value
        if weekday in fill_colors:
            color = fill_colors[weekday]
            for target_cell in ws[cell.coordinate[0] + str(cell.row - 1):cell.coordinate[0] + str(cell.row)]:
                target_cell.fill = PatternFill(start_color=color, end_color=color, fill_type="solid")

# Save formatted file
wb.save(file_name)
print(f"✅ Excel file '{file_name}' created with weekday-based color formatting.")
```
